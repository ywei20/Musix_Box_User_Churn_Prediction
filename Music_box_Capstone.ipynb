## 3. Label and Feature Generation

### 3.1 Label Definition

The purpose of this study is to model and predict churn possibility of users, therefore we need to create churn label. 

First set label window and feature window to be 14 and 30 days, respectively. Make sure they don't have overlaps, thus no leakage will happen.

All user id exist in feature windows will be stored in df_model_id. 

All user id exist in label windows will be stored in df_active_uid_in_label_window, these users are active users. 

Joining the above two variables yields churn users because their uid are missing from df_active_uid_in_label_window. In this way we can create a label column with '1' representing churn users and '0' representing active users.  

```python
import datetime
from dateutil import parser

label_window_size = 14
label_window_end_date = parser.parse('2017-05-12').date()
label_window_start_date = label_window_end_date - datetime.timedelta(label_window_size - 1)
print('label window:',label_window_start_date,'~',label_window_end_date,'days:',label_window_size)

feature_window_size = 30
feature_window_end_date = label_window_start_date - datetime.timedelta(1)
feature_window_start_date = feature_window_end_date  - datetime.timedelta(feature_window_size - 1)
print('feature window:',feature_window_start_date,'~',feature_window_end_date,'days:',feature_window_size)
```

```python
# all the uid we will model
df_model_uid = df.filter((F.col('date')>=feature_window_start_date) & (F.col('date')<=feature_window_end_date))\
                    .select('uid').distinct()
# active in label window (active label=0)
df_active_uid_in_label_window = df.filter((F.col('date')>=label_window_start_date) & (F.col('date')<=label_window_end_date))\
                            .select('uid').distinct().withColumn('label',F.lit(0))
```

```python
# prepare label data (churn label=1; active label=0)
df_label = df_model_uid.join(df_active_uid_in_label_window,on=['uid'],how='left')
df_label = df_label.fillna(1)
```

#### 3.2.1 Frequency features

For each user id, we can count event types 'P'/'D'/'S' over different time windows (e.g. 1d, 3d, 7d, 14d, 30d). For simpliciy, we write a function to achieve this target and store the result in 'df_frequency_list'.  

```python
def frequency_feature_generation_time_windows(df,event,time_window_list,snapshot_date):
    """
    generate frequency features for one event type and a list of time windows
    """
    df_frequency = df \
        .filter(F.col('event')==event) \
        .groupBy('uid') \
        .agg(*[F.sum(F.when((F.col('date')>=snapshot_date-datetime.timedelta(time_window-1)) & (F.col('date')<=snapshot_date),1).otherwise(0)).alias('freq_'+event+'_last_'+str(time_window)) \
                for time_window in time_window_list]
            )
    return df_frequency
```

We try to generate frequency feature on one event type, just to have an idea what the result table looks like.

```python
event = 'P'
time_window_list = [1,3,7,14,30]
snapshot_date = feature_window_end_date

df_frequency = frequency_feature_generation_time_windows(df_feature_window,event,time_window_list,snapshot_date)
df_frequency.show(5)
```

Now generate frequency features for all event_list, time_window_list

```python
event_list = ['P','D','S']
time_window_list = [1,3,7,14,30]
df_frequency_list = []
snapshot_date = feature_window_end_date
for event in event_list:
    df_frequency_list.append(frequency_feature_generation_time_windows(df_feature_window,event,time_window_list,snapshot_date))
```

#### 3.2.2 Recency feature

Recency feature is defined as days from last event; for each type of event, we can generate one feature.

Define a function to generate recency features. 

```python
from pyspark.sql.functions import datediff, lit
def recency_feature_generation(df, event, snapshot_date):
    df_recency = df.filter(F.col('event') == event).groupBy('uid').agg(F.max('date').alias('last_date')) \
                        .withColumn('days_from_last_'+event, datediff(F.lit(snapshot_date), 'last_date')) \
                        .drop('last_date')               
    return df_recency
```

```python
fancier_frequency_tmp = play_ds_tmp.filter(F.col('song_length') > 0) \
                                    .withColumn('song_pct', F.col('play_time')/F.col('song_length'))
fancier_frequency_tmp.show(5)
```

Create a new column 'if_song_played_80%' to represent whether a song is played no less than 80%. Note that if 'song_pct' has value over 1.0, it doesn't make sense. We only mark it as 1 when 'song_pct' is between 0.8 and 1.0, otherwise put 0.   

```python
fancier_frequency_tmp = play_ds_tmp.filter(F.col('song_length') > 0) \
                                    .withColumn('song_pct', F.col('play_time')/F.col('song_length')) \
                                    .withColumn('if_song_played_80%', F.when((F.col('song_pct') >= 0.8) \
                                                                             & (F.col('song_pct') <= 1.0), 1).otherwise(0)) \
                                    .select('uid','date', 'if_song_played_80%')
fancier_frequency_tmp.show(5)
```

Define a function to count songs played more than 80% over specified time window by each user. 

```python
def fancier_frequency_generation(df, time_window_list, snapshot_date):
    df_fancier_freq = df.groupBy('uid') \
                    .agg(*[F.sum(F.when((F.col('date')>= snapshot_date - datetime.timedelta(time_window -1)) & \
                                        (F.col('date') <= snapshot_date),F.col('if_song_played_80%')).otherwise(0)) \
                           .alias('play80%_freq_last_'+str(time_window)) for time_window in time_window_list]
                        )
    return df_fancier_freq
```

Generate counts of songs played more than 80% for different time windows using dataframe 'fancier_frequency_tmp' prepared a little earlier and store results in dataframe 'df_fancier_freq'.

```python
time_window_list = [1,3,7,14,30]
snapshot_date = feature_window_end_date

df_fancier_freq = fancier_frequency_generation(fancier_frequency_tmp, time_window_list, snapshot_date)
df_fancier_freq.show(5)
```

#### Evaluate Performance

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, roc_auc_score

def plot_roc_curve(y_train, y_train_pred, y_test, y_test_pred):
    roc_auc_train = roc_auc_score(y_train, y_train_pred)
    fpr_train, tpr_train, _ = roc_curve(y_train, y_train_pred)

    roc_auc_test = roc_auc_score(y_test, y_test_pred)
    fpr_test, tpr_test, _ = roc_curve(y_test, y_test_pred)
    plt.figure()
    lw = 2
    plt.plot(fpr_train, tpr_train, color='green',
             lw=lw, label='ROC Train (AUC = %0.4f)' % roc_auc_train)
    plt.plot(fpr_test, tpr_test, color='darkorange',
             lw=lw, label='ROC Test (AUC = %0.4f)' % roc_auc_test)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()
    
```

An ROC curve is the most commonly used way to visualize the performance of a binary classifier. So we plot ROC curves for models of training set and test set separately and present their corresponding AUC values. AUC on test is 0.8873, slightly better than AUC on train 0.8815, implying that logistic regression works quite good in churn prediction of the dataset. 

```python
y_train = res_train['label']
y_train_pred = [v[1] for v in res_train['probability']]
y_test = res_test['label']
y_test_pred = [v[1] for v in res_test['probability']]

plot_roc_curve(y_train, y_train_pred, y_test, y_test_pred)
```
